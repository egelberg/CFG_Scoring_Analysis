# All Data

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.options.display.float_format = '{:.2f}'.format
```

```python
import json

with open('Women/2016_games.json') as f:
    data = json.load(f)
    
data.keys()
```

```python
leaderboard = data['leaderboardRows']
```

```python
leaderboard
```

```python
database = []
for i in range(len(leaderboard)):
    competitors = leaderboard[i]['entrant']['competitorName']
    for j in range(len(leaderboard[i]['scores'])):
        workout_number = leaderboard[i]['scores'][j]['ordinal']
        times = leaderboard[i]['scores'][j]['time']
        ranks = leaderboard[i]['scores'][j]['rank']
        score = leaderboard[i]['scores'][j]['breakdown']
        database.append([competitors, workout_number, times, ranks, score])
        
```

```python
full_data = pd.DataFrame(database, columns=['competitors', 'ordinals', 'times', 'ranks', 'score'])
```

```python
full_data
```

# Event 1

```python
event_1 = full_data[full_data['ordinals']==1].sort_values('times')
```

```python
from datetime import datetime
total_seconds = []
for i, j in event_1.iterrows():
    if 'CAP' in j['times']:
        total_seconds.append(60*60)
    else:
        datetime_convert = datetime.strptime(j['times'],'%M:%S.%f')
        total_seconds.append(datetime_convert.second + datetime_convert.minute*60)
```

```python
event_1['total_seconds'] = total_seconds

lst = list(event_1['total_seconds']-min(event_1['total_seconds']))
event_1['adjusted_score'] = [100 - i*(100/max(lst)) for i in lst]
```

```python
event_1.sort_values('adjusted_score', ascending=False)
```

# Final Analysis

```python
frames = [event_1, event_2, event_3, event_4, event_5, event_6, event_7, event_8, event_9, event_10, event_11, event_12, event_13, event_14, event_15]
final_df = pd.concat(frames)[['competitors','adjusted_score', 'score']]
final_df['score'] = [int(i[:-4]) for i in final_df['score']]
```

```python
adjusted_placements = final_df.groupby('competitors').agg({'adjusted_score':'sum','score':'sum'}).sort_values('adjusted_score', ascending=False)
```

```python
adjusted_placements['adjusted_placement'] = [i+1 for i in range(len(adjusted_placements))]
```

```python
actual_placement = []

for i in range(len(leaderboard)):
    competitors = leaderboard[i]['entrant']['competitorName']
    placement = leaderboard[i]['overallRank']
    actual_placement.append((competitors, placement))
```

```python
final_placements = pd.DataFrame(actual_placement, columns=['competitors','actual_placement'])
```

```python
final_placements
```

```python
df = final_placements.join(adjusted_placements, how='inner', on='competitors')
df = df[['competitors','score','actual_placement','adjusted_score','adjusted_placement']]
```
