# All Data

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
import json

with open('Women/2020_games.json') as f:
    data = json.load(f)
    
data.keys()
```

```python
leaderboard = data['leaderboardRows']
```

```python
leaderboard
```

```python
database = []
for i in range(len(leaderboard)):
    competitors = leaderboard[i]['entrant']['competitorName']
    for j in range(len(leaderboard[i]['scores'])):
        workout_number = leaderboard[i]['scores'][j]['ordinal']
        times = leaderboard[i]['scores'][j]['time']
        ranks = leaderboard[i]['scores'][j]['rank']
        score = leaderboard[i]['scores'][j]['breakdown']
        database.append([competitors, workout_number, times, ranks, score])
        
```

```python
full_data = pd.DataFrame(database, columns=['competitors', 'ordinals', 'times', 'ranks', 'score'])
```

```python
full_data
```

# Event 1

```python
event_1 = full_data[full_data['ordinals']==1].sort_values('times')
```

```python
total_seconds = []
for i, j in event_1.iterrows(): 
    datetime_convert = datetime.strptime(j['times'],'%M:%S')
    total_seconds.append(datetime_convert.second + datetime_convert.minute*60)
```

```python
event_1['total_seconds'] = total_seconds

lst = list(event_1['total_seconds']-min(event_1['total_seconds']))
event_1['adjusted_score'] = [100 - i*(100/max(lst)) for i in lst]
```

```python
event_1
```

# Event 6

```python
event_6 = full_data[full_data['ordinals']==6].sort_values('times')
```

```python
total_seconds = []
for i, j in event_6.iterrows(): 
    try:
        datetime_convert = datetime.strptime(j['times'],'%M:%S')
        total_seconds.append(datetime_convert.second + datetime_convert.minute*60)
    except:
        datetime_convert = datetime.strptime(j['times'],':%S')
        total_seconds.append(datetime_convert.second + datetime_convert.minute*60)
```

```python
event_6['total_seconds'] = total_seconds

lst = list(event_6['total_seconds']-min(event_6['total_seconds']))
event_6['adjusted_score'] = [i*(100/max(lst)) for i in lst]
```

```python
event_6 = event_6.sort_values('adjusted_score', ascending=False)
event_6
```

```python
%matplotlib
import numpy as np

fig, ax = plt.subplots()

names = list(event_6['competitors'])
y_pos = np.arange(len(names))
performance = list(event_6['adjusted_score'])

ax.barh(y_pos, performance)
ax.set_yticks(y_pos)
ax.set_yticklabels(names)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Points')
ax.set_title('Crossfit Games 2020 Women Event 6: Max Handstand Hold')
print(y_pos[0], performance[0])

plt.barh(y_pos[0], performance[0], color=['red'])
plt.text(performance[0]+2, 0, '2:54, 100 pts', color='black', fontweight='bold', fontsize=10, ha='left', va='center')
plt.barh(y_pos[1], performance[1], color=['red'])
plt.text(performance[1]+2, 1, '2:13, 70.71 pts', color='black', fontweight='bold', fontsize=10, ha='left', va='center')
plt.barh(y_pos[2], performance[2], color=['red'])
plt.text(performance[2]+2, 2, '2:12, 70.00 pts', color='black', fontweight='bold', fontsize=10, ha='left', va='center')
plt.barh(y_pos[-1], performance[-1], color=['red'])
plt.text(performance[-1]+2, 29, '0:34, 0 pts', color='black', fontweight='bold', fontsize=10, ha='left', va='center')

fig.subplots_adjust(left=.25)
plt.xlim(0,performance[0]+20)
plt.show()
```

# Final Analysis

```python
frames = [event_1, event_2, event_3, event_4, event_5, event_6, event_7]
final_df = pd.concat(frames)[['competitors','adjusted_score', 'score']]
final_df['score'] = [int(i[:-4]) for i in final_df['score']]
```

```python
adjusted_placements = final_df.groupby('competitors').agg({'adjusted_score':'sum','score':'sum'}).sort_values('adjusted_score', ascending=False)
```

```python
adjusted_placements['adjusted_placement'] = [i+1 for i in range(len(adjusted_placements))]
```

```python
actual_placement = []

for i in range(len(leaderboard)):
    competitors = leaderboard[i]['entrant']['competitorName']
    placement = leaderboard[i]['overallRank']
    actual_placement.append((competitors, placement))
```

```python
final_placements = pd.DataFrame(actual_placement, columns=['competitors','actual_placement'])
```

```python
final_placements
```

```python
df = final_placements.join(adjusted_placements, how='inner', on='competitors')
df = df[['competitors','score','actual_placement','adjusted_score','adjusted_placement']]
```
