```python
import pandas as pd
```

```python
Men2015 = pd.read_csv('Adjusted_Results/Men_2015.csv', index_col='Unnamed: 0')
Men2016 = pd.read_csv('Adjusted_Results/Men_2016.csv', index_col='Unnamed: 0')
Men2017 = pd.read_csv('Adjusted_Results/Men_2017.csv', index_col='Unnamed: 0')
Men2018 = pd.read_csv('Adjusted_Results/Men_2018.csv', index_col='Unnamed: 0')
Men2019 = pd.read_csv('Adjusted_Results/Men_2019.csv', index_col='Unnamed: 0')
Men2020 = pd.read_csv('Adjusted_Results/Men_2020.csv', index_col='Unnamed: 0')
```

```python
Women2015 = pd.read_csv('Adjusted_Results/Women_2015.csv', index_col='Unnamed: 0')
Women2016 = pd.read_csv('Adjusted_Results/Women_2016.csv', index_col='Unnamed: 0')
Women2017 = pd.read_csv('Adjusted_Results/Women_2017.csv', index_col='Unnamed: 0')
Women2018 = pd.read_csv('Adjusted_Results/Women_2018.csv', index_col='Unnamed: 0')
Women2019 = pd.read_csv('Adjusted_Results/Women_2019.csv', index_col='Unnamed: 0')
Women2020 = pd.read_csv('Adjusted_Results/Women_2020.csv', index_col='Unnamed: 0')
```

```python
Diff_Men2015 = Men2015[Men2015.actual_placement != Men2015.adjusted_placement]
Diff_Men2015['difference'] = Diff_Men2015.actual_placement - Diff_Men2015.adjusted_placement 

Diff_Men2016 = Men2016[Men2016.actual_placement != Men2016.adjusted_placement]
Diff_Men2016['difference'] = Diff_Men2016.actual_placement - Diff_Men2016.adjusted_placement 

Men2017['actual_placement'] = pd.to_numeric(Men2017['actual_placement'], errors='coerce')
Men2017 = Men2017.dropna(subset=['actual_placement'])
Men2017['actual_placement'] = [int(i) for i in Men2017.actual_placement]
Diff_Men2017 = Men2017[Men2017.actual_placement != Men2017.adjusted_placement]
Diff_Men2017['difference'] = Diff_Men2017.actual_placement - Diff_Men2017.adjusted_placement

Diff_Men2018 = Men2018[Men2018.actual_placement != Men2018.adjusted_placement]
Diff_Men2018['difference'] = Diff_Men2018.actual_placement - Diff_Men2018.adjusted_placement 

Men2019['actual_placement'] = pd.to_numeric(Men2019['actual_placement'], errors='coerce')
Men2019 = Men2019.dropna(subset=['actual_placement'])
Men2019['actual_placement'] = [int(i) for i in Men2019.actual_placement]
Diff_Men2019 = Men2019[Men2019.actual_placement != Men2019.adjusted_placement]
Diff_Men2019['difference'] = Diff_Men2019.actual_placement - Diff_Men2019.adjusted_placement

Diff_Men2020 = Men2020[Men2020.actual_placement != Men2020.adjusted_placement]
Diff_Men2020['difference'] = Diff_Men2020.actual_placement - Diff_Men2020.adjusted_placement 
```

```python
Diff_Women2015 = Women2015[Women2015.actual_placement != Women2015.adjusted_placement]
Diff_Women2015['difference'] = Diff_Women2015.actual_placement - Diff_Women2015.adjusted_placement 

Diff_Women2016 = Women2016[Women2016.actual_placement != Women2016.adjusted_placement]
Diff_Women2016['difference'] = Diff_Women2016.actual_placement - Diff_Women2016.adjusted_placement 

Women2017['actual_placement'] = pd.to_numeric(Women2017['actual_placement'], errors='coerce')
Women2017 = Women2017.dropna(subset=['actual_placement'])
Women2017['actual_placement'] = [int(i) for i in Women2017.actual_placement]
Diff_Women2017 = Women2017[Women2017.actual_placement != Women2017.adjusted_placement]
Diff_Women2017['difference'] = Diff_Women2017.actual_placement - Diff_Women2017.adjusted_placement

Diff_Women2018 = Women2018[Women2018.actual_placement != Women2018.adjusted_placement]
Diff_Women2018['difference'] = Diff_Women2018.actual_placement - Diff_Women2018.adjusted_placement 

Women2019['actual_placement'] = pd.to_numeric(Women2019['actual_placement'], errors='coerce')
Women2019 = Women2019.dropna(subset=['actual_placement'])
Women2019['actual_placement'] = [int(i) for i in Women2019.actual_placement]
Diff_Women2019 = Women2019[Women2019.actual_placement != Women2019.adjusted_placement]
Diff_Women2019['difference'] = Diff_Women2019.actual_placement - Diff_Women2019.adjusted_placement

Diff_Women2020 = Women2020[Women2020.actual_placement != Women2020.adjusted_placement]
Diff_Women2020['difference'] = Diff_Women2020.actual_placement - Diff_Women2020.adjusted_placement 
```
